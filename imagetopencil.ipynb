```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
image = cv2.imread("download.jpeg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(image)
plt.axis("off")
plt.title("Uploaded Image")
plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(5,5))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("Gray_Scale Image")
plt.show()
```

```python
image_invert = cv2.bitwise_not(image_gray)
plt.figure(figsize=(5,5))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(5,5))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(image_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(10,10))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Uploaded Image")
plt.subplot(1,5,2)
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
